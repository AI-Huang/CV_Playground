```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
# @Date    : Nov-27-22 00:05
# @Author  : Kan HUANG (kan.huang@connect.ust.hk)


import numpy as np
import sys
sys.path.append("../..")
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
import matplotlib.pyplot as plt
from data_loaders.tf_fn.load_cifar10 import load_cifar10
from data_loaders.tf_fn.data_generator import get_datagenerator, get_datagenerator_constant
```

```python
# Load the CIFAR10 data.
(x_train, y_train), (x_val, y_val), (x_test, y_test) = load_cifar10(to_categorical=False)
print('Using keras augmentation.')
datagen = get_datagenerator(x=x_train)
datagen_constant = get_datagenerator_constant(x=x_train)
```

```python
iterator = datagen.flow(x_train, y_train, batch_size=64, shuffle=False)
iterator_constant = datagen_constant.flow(x_train, y_train, batch_size=64, shuffle=False)
```

## Nearest padding

```python
(x, y) = iterator.next()
train_images, train_labels = x, y

# https://www.tensorflow.org/tutorials/images/cnn?hl=zh-cn

class_names = ['airplane', 'automobile', 'bird', 'cat',
               'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i+1)
    plt.xticks(np.arange(0, 35, 5))
    plt.yticks(np.arange(0, 35, 5))
    plt.grid(True)
    plt.imshow(train_images[i])
    # The CIFAR labels happen to be arrays,
    # which is why you need the extra index
    plt.xlabel(class_names[train_labels[i][0]])
plt.show()
```
